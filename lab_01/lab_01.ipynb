** 1. Сколько мужчин и женщин (признак sex) представлено в этом наборе данных? **

** 2. Каков средний возраст (признак age) женщин? **

```python
female = data[data['sex'] == 'Female']
average = round(female['age'].mean(), 2) #округление до 2 знаков после точки
print(average)
```

Ответ: средний возраст женщин = 36.86

** 3. Какова доля граждан Германии (признак native-country)? **

```python
obs =  data['native-country'].count() * 0.01
germanyPeople = data[data['native-country'] == 'Germany']['age'].count()
                   
print(round(germanyPeople / obs, 2))
```

Ответ: доля граждан Германии = 0.42%

** 4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак salary) и тех, кто получает менее 50K в год? **

```python
poor = data[(data['salary'] == '<=50K')]['age'].mean()
rich = data[(data['salary'] == '>50K')]['age'].mean()
print(round(poor, 2))
print(round(rich, 2))

poor = data[(data['salary'] == '<=50K')]['age'].std()
rich = data[(data['salary'] == '>50K')]['age'].std()
print(round(poor, 2))
print(round(rich, 2))
```

Ответ:
- те, кто получает более 50К в год: средние значения = 44.25, среднеквадратичные отклонения возраста = 10.52
- те, кто получает менее 50К в год: средние значения = 36.78, среднеквадратичные отклонения возраста = 14.02

** 7. Выведите статистику возраста для каждой расы (признак race) и каждого пола. Используйте groupby и describe. Найдите таким образом максимальный возраст мужчин расы Amer-Indian-Eskimo. **

** 8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми. **

```python
obc = data[data['sex'] == 'Male']['sex'].count() * 0.01
married = data[(data['salary'] == '>50K') & (data['sex'] == 'Male') & (data['marital-status'].apply( \
                lambda x: 'married-' in x.lower()))]['sex'].count()
unmarried = data[(data['salary'] == '>50K') & (data['sex'] == 'Male') & (~data['marital-status'].apply( \
                lambda x: 'married-' in x.lower()))]['sex'].count()

print(round(married / obc, 2))
print(round(unmarried / obc, 2))
```

То же самое, но реализовано в виде списка:

```python
list = ['Married-civ-spouse', 'Married-spouse-absent' , 'Married-AF-spouse']
data['is_married'] = 0
data['is_married'] = (data['marital-status'].isin(list)) * 1
round(pd.crosstab(data['salary'], data['is_married'], normalize = True) * 100, 2)
```

** 9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много? **

```python
hoursCount = data['hours-per-week'].max() 
print(hoursCount)

count = data[data['hours-per-week'] == hoursCount].shape[0]
print(count)

richCount = data[data['hours-per-week'] == hoursCount]['salary'].value_counts(normalize='>50K')
print(richCount)
```

Ответ: максимальное число часов в неделю = 99, количество часов = 85, зарабатываюших много = 0.294118
