```python
import pandas as pd
import os

from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import graph_objs as go
import requests
import pandas as pd

print(__version__) # need 1.9.0 or greater

init_notebook_mode(connected = True)

def plotly_df(df, title = ''):
    data = []
    
    for column in df.columns:
        trace = go.Scatter(
            x = df.index,
            y = df[column],
            mode = 'lines',
            name = column
        )
        data.append(trace)
    
    layout = dict(title = title)
    fig = dict(data = data, layout = layout)
    iplot(fig, show_link=False)
```

## Предсказание с помощью Facebook Prophet

Для начала построим предсказание с помощью простой библиотеки `Facebook Prophet`. Для того, чтобы посмотреть на качество модели, отбросим из обучающей выборки последние 30 дней.

```python
from fbprophet import Prophet
```

```python
predictions = 30

df = df[['date', 'count']]
df.columns = ['ds', 'y']
train_df = df[:-predictions].copy()
```

```python
df.tail()
```

```python
prophetObj = Prophet(yearly_seasonality=False, daily_seasonality=False)
```

```python
prophetObj.fit(train_df)
```

```python
makeFutureDF = prophetObj.make_future_dataframe(periods=predictions)
```

```python
forecast = prophetObj.predict(makeFutureDF)
forecast.tail()
```

**Вопрос 1:** Какое предсказание числа просмотров wiki-страницы на 20 января? Ответ округлите до целого числа.

Оценим качество предсказания по последним 30 точкам. <br>
**Ответ 3428**

```python
round(forecast['yhat'][-1:])
```

```python
import numpy as np

cmpDF = forecast.set_index('ds')[['yhat', 'yhat_lower', 'yhat_upper']].join(df.set_index('ds'))
cmpDF['e'] = cmpDF['y'] - cmpDF['yhat']
cmpDF['p'] = 100 * cmpDF['e'] / cmpDF['y']
print(np.mean(abs(cmpDF[-predictions:]['p'])))
print(np.mean(abs(cmpDF[-predictions:]['e'])))
```

**Вопрос 5**: Далее перейдем к построению модели SARIMAX (`sm.tsa.statespace.SARIMAX`). Модель c какими параметрами лучшая по `AIC`-критерию?<br>
**Ответ лучшая модель SARIMAX(1, 0, 2)x(3, 0, 0, 7) и AIC 5020.963**

```python
p = 1
q = 2
P = 3
ds = range(0, 2)
Ds = range(0, 3)
Qs = range(0, 2)
```

```python
from itertools import product
parameters = product(Ds, ds, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
from tqdm import tqdm
results = []
best_aic = float("inf")



for param in tqdm(parameters_list):
    try:
        model=sm.tsa.statespace.SARIMAX(train_df['y'], order=(p, param[1], q), seasonal_order=(P, param[0], param[2], 7)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
result = pd.DataFrame(results)
result.columns = ['parameters', 'aic']
print(result.sort_values(by = 'aic', ascending=True).head())
```
